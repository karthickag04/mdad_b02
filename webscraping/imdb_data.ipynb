```python




movies = []

# ----------------------------------
# Iterate over each movie block
# ----------------------------------
for item in soup.select("li.ipc-metadata-list-summary-item"):
    
    # Title
    title_tag = item.select_one("h3.ipc-title__text")
    title = title_tag.get_text(strip=True) if title_tag else None

    # Year, Duration, Certificate
    metadata = item.select(".dli-title-metadata-item")
    year = metadata[0].text if len(metadata) > 0 else None
    duration = metadata[1].text if len(metadata) > 1 else None
    certificate = metadata[2].text if len(metadata) > 2 else None

    # IMDb Rating & Votes
    rating_tag = item.select_one(".ipc-rating-star--rating")
    rating = rating_tag.text if rating_tag else None

    votes_tag = item.select_one(".ipc-rating-star--voteCount")
    votes = votes_tag.text.strip("() ") if votes_tag else None

    # Plot
    plot_tag = item.select_one(".title-description-plot-container .ipc-html-content-inner-div")
    plot = plot_tag.text.strip() if plot_tag else None

    # Director
    director_tag = item.select_one("span:contains('Director') + span a")
    director = director_tag.text if director_tag else None

    # Stars
    stars = [a.text for a in item.select("span:contains('Stars') + span a")]

    # Poster Image
    img_tag = item.select_one("img.ipc-image")
    poster_url = img_tag["src"] if img_tag else None

    # Extra Info Box (Release, Budget, Collections, Verdict)
    extra_info = {}
    info_block = item.select_one("[data-testid='title-list-item-description']")
    if info_block:
        lines = info_block.get_text(separator="\n").split("\n")
        for line in lines:
            if ":" in line:
                k, v = line.split(":", 1)
                extra_info[k.strip()] = v.strip()

    movies.append({
        "Title": title,
        "Year": year,
        "Duration": duration,
        "Certificate": certificate,
        "IMDb_Rating": rating,
        "Votes": votes,
        "Director": director,
        "Stars": ", ".join(stars),
        "Plot": plot,
        "Poster_URL": poster_url,
        **extra_info
    })



print(movies)
for movie in movies:
    print(movie)
    # for key, value in movie.items():
    #     print(f"  {key}: {value}")

# ----------------------------------
# Convert to Pandas DataFrame
# ----------------------------------
# df = pd.DataFrame(movies)

# print(df)
```
