## Feature Encoding

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('data/EDA_01.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(columns = ["CustomerId", "RowNumber", "Surname"], axis = 1, inplace=True)
```

```python
df.head(5)
```

```python
df.Gender.value_counts()
```

#### Handling missing values in Gender

```python
df.Gender.mode()
```

```python
df['Gender'] = df['Gender'].fillna('Male')
```

```python
df['Age'] = df['Age'].fillna(df.Age.mean())
```

```python
df.Gender.value_counts()
```

```python
df.info()
```

#### Label Encoding

```python
le = preprocessing.LabelEncoder()
df['Gender_label'] = le.fit_transform(df.Gender.values)
```

```python
df.sample(5)
```

```python
df.Gender_label.value_counts()
```

#### One Hot Encoding

```python
one_hot = pd.get_dummies(df['Geography'])
one_hot
```

```python
one_hot=one_hot.astype(int)
one_hot
```

```python
df_dummies = pd.get_dummies(df)
df_dummies = df_dummies.astype(int)
df_dummies.head(5)
```
