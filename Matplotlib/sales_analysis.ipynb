# Sales Analysis
This notebook provides a comprehensive sales analysis using the dataset `dataset_supermarket_sales.csv`.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('sales_data.csv')

# 'yyyy-mm-dd'

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# Extract hour from 'Time'
df['Hour'] = df['Time'].apply(lambda x: int(x.split(':')[0]))
df['Invoice ID01'] = df['Invoice ID'].apply(lambda x: int(x.split('-')[0]))

df.head(5)
```

## 1. Descriptive Statistics

```python
# Total Sales
total_sales = df['Total'].sum()
print(total_sales)
```

```python
# Average Sales
average_sales = df['Total'].mean()

print(average_sales)
```

```python
# Monthly Sales Trend
monthly_sales_trend = df.resample('M', on='Date')['Total'].sum()
print(monthly_sales_trend)
```
