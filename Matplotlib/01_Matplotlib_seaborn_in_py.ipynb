```python
import pandas as pd
import matplotlib.pyplot as plt

def format_func(value, total):
    return f'{int(value)}'

# Sample data
df = pd.DataFrame({'Group': ['Group 1', 'Group 2', 'Group 3', 'Group 4'],
                   'Size': [15, 30, 45, 10]})
plt.pie(df['Size'], labels=df['Group'], autopct=lambda p: format_func(p * sum(df['Size']) / 100, sum(df['Size'])), colors=['red', 'blue', 'green', 'yellow'])
plt.title('Pie Chart Example')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Surface Plot Example')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Sample data
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Create a 3D scatter plot with animation
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(size=5, color=z, colorscale='Viridis', opacity=0.8)
))

# Adding animation frames
frames = [go.Frame(data=[go.Scatter3d(
    x=x[:k],
    y=y[:k],
    z=z[:k],
    mode='markers',
    marker=dict(size=5, color=z[:k], colorscale='Viridis', opacity=0.8)
)]) for k in range(1, len(t))]

fig.frames = frames

# Add animation settings
fig.update_layout(
    updatemenus=[dict(type="buttons", showactive=False,
                      buttons=[dict(label="Play",
                                    method="animate",
                                    args=[None, {"frame": {"duration": 50, "redraw": True},
                                                 "fromcurrent": True, "mode": "immediate"}])])])

fig.update_layout(scene=dict(
                    xaxis_title='X Axis',
                    yaxis_title='Y Axis',
                    zaxis_title='Z Axis'),
                  title="Dynamic 3D Scatter Plot with Animation")

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Sample data for the animation
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 20, 30, 40]

# Create initial figure
fig = go.Figure()

# Create initial donut chart
fig.add_trace(go.Pie(
    labels=labels, 
    values=values, 
    hole=0.5, 
    textinfo='label+percent', 
    insidetextorientation='radial',
    hoverinfo='label+value',
    marker=dict(colors=['#ff9999','#66b3ff','#99ff99','#ffcc99']),
))

# Create animation frames by modifying values
frames = []
for i in range(10):
    values = np.random.randint(10, 50, size=4)  # Randomly changing values
    frames.append(go.Frame(data=[go.Pie(labels=labels, values=values, hole=0.5)]))

# Add frames to the figure
fig.frames = frames

# Add animation settings
fig.update_layout(
    updatemenus=[dict(type="buttons", showactive=False,
                      buttons=[dict(label="Play",
                                    method="animate",
                                    args=[None, {"frame": {"duration": 500, "redraw": True},
                                                 "fromcurrent": True, "mode": "immediate"}])])])

# Update layout for donut chart appearance
fig.update_traces(textposition='inside')
fig.update_layout(
    title_text="Animated Donut Chart Example",
    annotations=[dict(text='Donut Chart', x=0.5, y=0.5, font_size=20, showarrow=False)],
    showlegend=False
)

fig.show()
```
