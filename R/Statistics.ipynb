```r
# Load dataset
data(mtcars)

# View the first few rows of the dataset
head(mtcars)

# ------------------------------------------
# 1. Mean, Median, Mode
# ------------------------------------------

# Mean of 'mpg' (Miles per Gallon)
mean_mpg <- mean(mtcars$mpg)
print(paste("Mean MPG:", mean_mpg))

# Median of 'mpg'
median_mpg <- median(mtcars$mpg)
print(paste("Median MPG:", median_mpg))

# Mode (custom function because R doesn't have a built-in one)
get_mode <- function(x) {
  uniq_vals <- unique(x)
  uniq_vals[which.max(tabulate(match(x, uniq_vals)))]
}
mode_cyl <- get_mode(mtcars$cyl)
print(paste("Mode of Cylinders:", mode_cyl))

# ------------------------------------------
# 2. Min, Max, Range
# ------------------------------------------

# Minimum & Maximum
min_mpg <- min(mtcars$mpg)
max_mpg <- max(mtcars$mpg)

# Range (returns min and max)
range_mpg <- range(mtcars$mpg)

print(paste("Min MPG:", min_mpg))
print(paste("Max MPG:", max_mpg))
print(paste("Range of MPG:", paste(range_mpg, collapse = " - ")))

# ------------------------------------------
# 3. Quantiles & Interquartile Range
# ------------------------------------------

# Quantiles (25%, 50%, 75%)
quantiles <- quantile(mtcars$mpg)
print("Quantiles of MPG:")
print(quantiles)

# Interquartile Range (IQR)
iqr_mpg <- IQR(mtcars$mpg)
print(paste("Interquartile Range of MPG:", iqr_mpg))

# ------------------------------------------
# 4. Variance & Standard Deviation
# ------------------------------------------

# Variance
var_mpg <- var(mtcars$mpg)
# Standard Deviation
sd_mpg <- sd(mtcars$mpg)

print(paste("Variance of MPG:", var_mpg))
print(paste("Standard Deviation of MPG:", sd_mpg))

# ------------------------------------------
# 5. Summary Statistics
# ------------------------------------------

# Summary of a specific column
summary(mtcars$mpg)

# Summary of the entire dataset
summary(mtcars)

# Five-number summary (Tukey)
fivenum_mpg <- fivenum(mtcars$mpg)
print("Five-number summary of MPG:")
print(fivenum_mpg)

# ------------------------------------------
# 6. Frequency Table
# ------------------------------------------

# Frequency table of cylinder counts
cyl_table <- table(mtcars$cyl)
print("Frequency of Cylinder Values:")
print(cyl_table)

# Relative frequencies
rel_freq <- prop.table(cyl_table)
print("Relative Frequencies:")
print(rel_freq)

# ------------------------------------------
# 7. Boxplot and Histogram (optional for visuals)
# ------------------------------------------

# Boxplot
boxplot(mtcars$mpg, main = "Boxplot of MPG", col = "lightblue")

# Histogram
hist(mtcars$mpg, main = "Histogram of MPG", col = "orange", breaks = 10)


```
