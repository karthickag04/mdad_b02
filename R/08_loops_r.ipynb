```r
# loops in R language

# 1. For loop with vector
numbers <- c(1, 2, 3, 4, 5)
print("For loop with vector:")
for (i in numbers) {
    print(paste("Number:", i))
}
```

```r

# 2. For loop with index
print("\nFor loop with index:")
for (i in 1:5) {
    print(paste("Index:", i, "Value:", numbers[i]))
}

```

```r

# 3. For loop with list
fruits <- list("apple", "banana", "orange")
print("\nFor loop with list:")
for (fruit in fruits) {
    print(paste("Fruit:", fruit))
}
```

```r

# 4. While loop
print("\nWhile loop:")
counter <- 1
while (counter <= 5) {
    print(paste("Counter:", counter))
    counter <- counter + 1
}
```

```r

# 5. Repeat loop with break
print("\nRepeat loop:")
x <- 1
repeat {
    print(paste("X value:", x))
    x <- x + 1
    if (x > 5) break
}
```

```r


# 6. Nested loops
print("\nNested loops - multiplication table:")
for (i in 1:3) {
    for (j in 1:3) {
        result <- i * j
        cat(paste(i, "x", j, "=", result, "  "))
    }
    cat("\n")
}
```

```r

# 7. Loop with data frame
df <- data.frame(name = c("John", "Jane", "Bob"), age = c(25, 30, 35))
print("\nLoop through data frame rows:")
for (i in 1:nrow(df)) {
    print(paste("Name:", df$name[i], "Age:", df$age[i]))
}
```

```r

# 8. Apply family functions (vectorized alternatives)
print("\nUsing lapply (alternative to loops):")
squared <- lapply(numbers, function(x) x^2)
print(squared)
```
