```r
x <- 1:10
y <- rnorm(10)

# Basic plot with some customizations
plot(x, y, 
     type = "o",             # Points and lines
     col = "blue",           # Blue color for points and lines
     pch = 16,               # Solid circle for points
     cex = 1.5,              # Increase point size
     lwd = 2,                # Line width = 2
     lty = 2,                # Dashed line
     main = "Customized Plot", # Main title
     xlab = "X Axis",        # X-axis label
     ylab = "Y Axis",        # Y-axis label
     xlim = c(0, 12),        # X-axis limits
     ylim = c(-3, 3),        # Y-axis limits
     cex.main = 1.5,         # Main title size (reduce for testing)
     cex.axis = 1.2          # Axis label size (reduce for testing)
)
```
